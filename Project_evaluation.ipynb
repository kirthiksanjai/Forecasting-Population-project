```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import numpy as np

# Define the number of folds for cross-validation
n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=True)

# Assuming you have multiple target variables named 'target_variable1', 'target_variable2', etc.
target_variables = [
    "p00_1200", "p10_1200", "p00u_1200", "p10u_1200", "p00r_1200", "p10r_1200",
    "p00_30", "p10_30", "p00u_30", "p10u_30", "p00r_30", "p10r_30",
    "p00_75", "p10_75", "p00u_75", "p10u_75", "p00r_75", "p10r_75",
    "p00_150", "p10_150", "p00u_150", "p10u_150", "p00r_150", "p10r_150",
    "p00_600", "p10_600", "p00u_600", "p10u_600", "p00r_600", "p10r_600",
    "p00_300", "p10_300", "p00u_300", "p10u_300", "p00r_300", "p10r_300"
]

# Iterate over the folds
for train_index, val_index in kf.split(df_combined):
    # Get the training and validation sets
    train_set = df_combined.iloc[train_index]
    val_set = df_combined.iloc[val_index]

    # Split the training set into features (X_train) and target variables (y_train)
    X_train = train_set.drop(target_variables, axis=1)
    y_train = train_set[target_variables]

    # Split the validation set into features (X_val) and target variables (y_val)
    X_val = val_set.drop(target_variables, axis=1)
    y_val = val_set[target_variables]

    # Initialize the linear regression model
    model = LinearRegression()

    # Fit the model on the training set
    model.fit(X_train, y_train)

    # Evaluate the model on the validation set
    y_pred = model.predict(X_val)

    # Calculate the evaluation metric (e.g., RMSE) for each target variable
    for i, target_variable in enumerate(target_variables):
        rmse = np.sqrt(mean_squared_error(y_val[target_variable], y_pred[:, i]))
        print("RMSE for", target_variable, ":", rmse)
```
